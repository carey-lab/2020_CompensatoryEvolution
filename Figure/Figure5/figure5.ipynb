```python
#prepare the density scatter figure
import matplotlib as mpl
def reverse_colourmap(cmap, name = 'my_cmap_r'):      
    reverse = []
    k = []   
    for key in cmap._segmentdata:    
        k.append(key)
        channel = cmap._segmentdata[key]
        data = []
        for t in channel:                    
            data.append((1-t[0],t[2],t[1]))            
        reverse.append(sorted(data))    
    LinearL = dict(zip(k,reverse))
    my_cmap_r = mpl.colors.LinearSegmentedColormap(name, LinearL) 
    return my_cmap_r
```

```python
#plot the figure for AllN>=10
my_cmap = reverse_colourmap(mpl.cm.RdBu)
from scipy.stats import gaussian_kde
plt.rcParams["figure.dpi"] = 300
plt.figure(figsize=(2.5,2.5))
ax = plt.subplot(111)
x = np.array(big10Data["FitnessCorrelation"].values)
y = np.array(big10Data["N1/(N1+N2)"].values)*100
xy = np.vstack([x,y])
z = gaussian_kde(xy)(xy)

# Sort the points by density, so that the densest points are plotted last
idx = z.argsort()
x, y, z = x[idx], y[idx], z[idx]
plt.scatter(x, y, c=z,s=1.5,cmap=
            my_cmap)

plt.xlabel("O.R.")
plt.ylabel("Correlation")

#add the colorbar
cbar =plt.colorbar(shrink=0.6)
cbar.ax.set_title('Density', fontsize=6)
cax = plt.gcf().axes[-1]
cax.tick_params(labelsize=4,length=2,width=0.5)
cbar.outline.set_linewidth(0.5)

for axis in ['bottom','left']:
    ax.spines[axis].set_linewidth(0.5)

ax = plt.gca()
ax.tick_params(width=0.5)
#the length of the ticks
ax.tick_params(axis='both',length=2)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

#plt.ylim([0,1])
plt.tight_layout()
#save the data 
#calculate the correlation between these two factors
```

```python
#make sure that for every single feature, if one feature can predict fitness, this feature can also predict compensation where feature can predict compenasation
feature_singleList = subset_featurePredictCompensationData["feature"].values
#write a function to return the correlation of featurepredictCompensation and N1/(N1+N2)
cutOff = 50
def returnRFeatureFitnessVSN1N2(feature_NameSingle,cutOff):
    #filter the data
    feature_Data = subset_featurePredictCompensationData[subset_featurePredictCompensationData["feature"] == feature_NameSingle]
    correlation_big50List = [ ]
    count_big50List = [ ]
    
    for single_correlation, single_count in zip(feature_Data["correlation"].values,
                                           feature_Data["sum_count_result"].values):
    
        single_countN1, single_countN2 = literal_eval(single_count)
        if( single_countN2+single_countN1) >= cutOff:
            count_big50List.append(single_countN1/(single_countN1 + single_countN2))
            correlation_big50List.append(single_correlation)

    big50DataInner = pd.DataFrame({"FitnessCorrelation":correlation_big50List,
                         "N1/(N1+N2)":count_big50List})
    R_inner, P_inner = stats.spearmanr(big50DataInner["FitnessCorrelation"].values,
                     big50DataInner["N1/(N1+N2)"].values)
    return R_inner, P_inner

R_inner_list = [ ]
P_inner_list = [ ]
for single_feature in feature_singleList:
    R_inner1, P_inner1 = returnRFeatureFitnessVSN1N2(single_feature, cutOff)
    R_inner_list.append(R_inner1)
    P_inner_list.append(P_inner1)
#make the data 
RP_data = pd.DataFrame({"R":R_inner_list,
                       "P":P_inner_list})
```

```python
#clean the data 
RP_data["feature"] = feature_singleList
print(RP_data.shape[0])

RP_dataClean = RP_data[ (RP_data["feature"] != 'FAUJ880111') & (RP_data["feature"] != 'RACS820112') \
                      & (RP_data["feature"] != 'CORJ870103') & (RP_data["feature"] != 'FAUJ880111') \
                      & (RP_data["feature"] != 'GEOR030101') & (RP_data["feature"] != 'GUYH850101') \
                      & (RP_data["feature"] != 'HUTJ700102') & (RP_data["feature"] != 'KARS160112') \
                      & (RP_data["feature"] != 'MIYS990105') & (RP_data["feature"] != 'PALJ810111') \
                      & (RP_data["feature"] != 'PONP800106') & (RP_data["feature"] != 'QIAN880139') \
                      & (RP_data["feature"] != 'RACS820112')]

RP_dataClean = RP_dataClean[RP_dataClean["P"]<0.05]
#plot the hist 
plt.figure(figsize=(5,4))
ax = plt.subplot(111)
plt.hist(RP_dataClean["R"].values, bins = 20 ,color = "silver", rwidth=0.9)

plt.xlabel("Pearson correlations bewteen\n% of genotypes rescued when AAindex->WT\nand the correlations between AAindex value\ndistance to WT and fitness"
          ,fontsize = 12)
plt.ylabel("# of physicochemical features\nX Positions(Each cell in Fig. 4C)",fontsize = 13)

for axis in ['bottom','left']:
    ax.spines[axis].set_linewidth(0.5)

ax = plt.gca()
ax.tick_params(width=0.5)
#the length of the ticks
ax.tick_params(axis='both',length=2)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

#plt.ylim([0,1])
plt.tight_layout()
#save the figure 
plt.savefig(savedata_dir + "SingleFeaturePredictCompensation.png", dpi = 300)
```
