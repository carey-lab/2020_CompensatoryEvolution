```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib

params = {'font.family': 'Arial','font.weight':'normal'} #set the basic parameters for figure
plt.rcParams["figure.dpi"] = 300
matplotlib.rcParams.update(params)
```

```python
readdata_dir = "D:/Develop/Compensatory-Evolution/CodeForGithub/File/"
savedata_dir = "D:/Develop/Compensatory-Evolution/CodeForGithub/Figure10/"
```

```python
#show that Supercompensation can significantly stabilize the local structure of Hisp
FP_data = pd.read_csv(readdata_dir + "SuperCompensationFile/SubBufferingAbility.tsv",sep = "\t")
```

```python
plt.rcParams["figure.dpi"] = 300
plt.figure(figsize=(3,3))
ax = plt.subplot(111)
plt.scatter(FP_data["FitnessImpactScoreRelativeWT"].values,
           FP_data["ddG_foldOpt"].values,color = "silver", edgecolors="k",
           linewidths=0.2)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
#modify the lindwidth of the lines 
for axis in ['bottom','left']:
    ax.spines[axis].set_linewidth(0.5)
    
    
ax = plt.gca()
ax.tick_params(width=0.5)
ax.tick_params( axis='both',length=2)

#add the label names
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)

plt.ylabel(r'$\Delta$$\Delta$G of'+ " the protein stability",fontsize = 14)
plt.xlabel("Fitness impact score\n(relative to WT)", fontsize = 14)

plt.tight_layout()
#save the figure
plt.plot([0,0],[-22,90],ls = "--",color= 'k',linewidth=1)
plt.plot([-20,10],[0,0],ls = "--",color= 'k',linewidth=1)
plt.ylim([-22,90])
plt.xlim([-20,10])
plt.savefig(savedata_dir + "SCStabilizeProteinStructure.png", dpi = 300)
```
