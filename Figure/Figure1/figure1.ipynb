```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib

params = {'font.family': 'Arial','font.weight':'normal'} #set the basic parameters for figure
plt.rcParams["figure.dpi"] = 300
matplotlib.rcParams.update(params)
```

```python
readdata_dir = "D:/Develop/Compensatory-Evolution/CodeForGithub/File/"
savedata_dir = "D:/Develop/Compensatory-Evolution/CodeForGithub/Figure1/"
```

```python
#load rescuability data 
rescuability_data = pd.read_csv(readdata_dir + "MLFile/MLPrepare_data.tsv",sep = "\t")
rescuability_data
```

```python
big0_small01N = 6269
equal0N = 21682
big01N = 3750
big0_small01Ratio = 6269/rescuability_data.shape[0]*100
equal0Ratio = 21682/rescuability_data.shape[0]*100
big01Ratio = 3750/rescuability_data.shape[0]*100
print(equal0Ratio, big0_small01Ratio, big01Ratio)
print(equal0N,big0_small01N,big01N)
```

```python
#plot the figure
fig = plt.figure(figsize=(2.5,1.8))
plt.rcParams['figure.dpi'] = 300
bar1 = [21682]
bar3 = [3750]
bar2 = [6269]
# Heights of bars1 + bars2
bars = np.add(bar1, bar2).tolist()
ax = plt.subplot(111)

ax.bar(0.75,bar1,color="silver",label="=0 (68%)",width=0.2)
ax.bar(0.75,bar2,bottom=bar1,color="royalblue",label=">0,<0.1 (20%)",width = 0.2)
ax.bar(0.75,bar3,bottom=bars,color="firebrick",label = ">= 0.1 (12%)", width = 0.2)

plt.yticks(fontsize=6)
plt.xlabel("Rescuability",fontsize=7)
plt.ylabel("# of unfit genotypes",fontsize=7)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
#modify the lindwidth of the lines 
for axis in ['bottom','left']:
    ax.spines[axis].set_linewidth(0.5)
    
#modify the lindth of the ticks
ax = plt.gca()
ax.tick_params(width=0.5)
#the length of the ticks
ax.tick_params(axis='both',length=2)

plt.legend(fontsize=5,facecolor="white",edgecolor="white",loc = "center left",bbox_to_anchor=(1.01, 1.0))
plt.xlim(0.5,1.0)
plt.xticks([])
fig.tight_layout()
plt.savefig(savedata_dir + "Rescuability_distribution.png",dpi = 300)
```
