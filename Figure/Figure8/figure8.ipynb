```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib

params = {'font.family': 'Arial','font.weight':'normal'} #set the basic parameters for figure
plt.rcParams["figure.dpi"] = 300
matplotlib.rcParams.update(params)
```

```python
readdata_dir = "D:/Develop/Compensatory-Evolution/CodeForGithub/File/"
savedata_dir = "D:/Develop/Compensatory-Evolution/CodeForGithub/Figure8/"
```

```python
#plot the skewness of A110D and K115E
#load the data 
D110_buffer = pd.read_csv(readdata_dir + "SuperCompensationFile/A110D.txt", sep = "\t")
Q110_buffer = pd.read_csv(readdata_dir + "SuperCompensationFile/A110Q.txt", sep = "\t")
```

```python
#plot the FE absolute difference of these two different FEs
#plot the hist plot
plt.rcParams["figure.dpi"] = 300
plt.figure(figsize=(2.5,1.5))
ax = plt.subplot(111)
plt.hist(-1*D110_buffer["abd_FE_minus_wt"].values,bins=50,color="silver",rwidth=0.95)

plt.xlabel("|∆f| in 110A(WT) - |∆f| in 110D",fontsize=12)
#plt.ylabel("# of substitutions in\ndifferent genetic backgrounds",fontsize=6)


ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

for axis in ['bottom','left']:
    ax.spines[axis].set_linewidth(0.5)
    
#modify the lindth of the ticks
ax = plt.gca()
ax.tick_params(width=0.5)
#the length of the ticks
ax.tick_params(axis='both',length=2)




#plt.xticks([-0.5,0.00,0.5],[-0.5,0.00,0.5],fontsize = 5)
plt.yticks([0,2500,5000,7500],[0,2500,5000,7500],fontsize = 10)
plt.xlim(-0.4,0.5)

plt.tight_layout()
plt.savefig(savedata_dir  + "A110d_dist.png",dpi=300)
```
